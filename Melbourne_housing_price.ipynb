```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
```

```python
df=pd.read_csv("Melbourne_housing_FULL.csv")
df
```

```python
df.info()
```

```python
df.nunique()
```

```python
col = ['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df = df[col]
df
```

```python
df.dropna(inplace=True)
df
```

```python
newdf=pd.get_dummies(df,dtype=int)
x=newdf.drop(columns=['Price'])
y=newdf['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33)
```

```python
# Lasso model
from sklearn.linear_model import Lasso
reg1=Lasso(alpha=50,max_iter=100,tol=0.1)
reg1.fit(x_train,y_train)
```

```python
reg1.score(x_train,y_train)
```

```python
reg1.score(x_test,y_test)
```

```python
#Ridge Reg
from sklearn.linear_model import Ridge
reg2=Ridge(alpha=50,max_iter=100,tol=0.1)
reg2.fit(x_train,y_train)
```

```python
reg2.score(x_train,y_train)
```

```python
reg2.score(x_test,y_test)
```

```python
#Linear Reg
from sklearn.linear_model import LinearRegression
reg3=LinearRegression()
reg3.fit(x_train,y_train)
```

```python
reg3.score(x_train,y_train)
```

```python
reg3.score(x_test,y_test)
```

```python
#We can either use Lasso or Ridge when overfitting case comes in Linear Regression
```
